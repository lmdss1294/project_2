```python
import datetime

result = []

a = df['날짜']
b = df['계획시간']

if len(a) == len(b):
    for i in range(len(a)):

        a_dt = datetime.datetime.strptime(str(a.iloc[i]), '%Y%m%d')

        b_time = b.iloc[i].split(' ')[-1]
        b_str = ' '.join(b.iloc[i].split(' ')[:-1])

        b_dt = datetime.datetime.strptime(b_time, '%I:%M:%S')

        if '오후' in b_str:
            b_dt += datetime.timedelta(hours=12)

        result.append(f'{a_dt.date()} {b_dt.time()}')

df['일시'] = result
```
